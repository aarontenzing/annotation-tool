### Visualise the image

```python
import cv2
from matplotlib import pyplot as plt
from handle_json import read_json

data = read_json()
data = data[0]
image = cv2.imread('data\\' + data['img_name'])
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

### Read the coordinates from the json file

```python
points = data['projection']
edges = (
            (0, 1),
            (1, 2),
            (2, 3),
            (3, 0),
            (4, 5),
            (5, 6),
            (6, 7),
            (7, 4),
            (0, 4),
            (1, 5),
            (2, 6),
            (3, 7)
        )

for i in range(len(points)):
    points[i][1] = image.shape[0] - points[i][1]
print(points)
```

```python
def drawline(image, points):
    img_copy = image.copy()
    for i in edges:
        x = points[i[0]]
        y = points[i[1]]
        cv2.line(img_copy, (x[0], x[1]) , (img_copy.shape[0]-y[0], img_copy.shape[0]-y[1]), (255,0,0), 4)
    plt.imshow(img_copy)
```

```python

def drawpoint(image, points):
    img_copy = image.copy()
    for i in points:
        cv2.circle(img_copy, (img_copy.shape[0]-i[0], img_copy.shape[0]-i[1]), 0, (255,0,0), 30)
    plt.imshow(img_copy)
```

```python
# Drawing the vertices and center of the box
drawpoint(image, points)
```

```python
# drawing all the lines
drawline(image, points)
```
