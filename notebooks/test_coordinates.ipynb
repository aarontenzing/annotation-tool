```python
import cv2
from matplotlib import pyplot as plt
from  handle_json import read_json

data = read_json('..\\annotations.json')

print(data[7]['img_name'])
print(data[7]['projection'])

image = cv2.imread('..\\data\\' + str(data[7]['img_name']))
w, h = image.shape[1], image.shape[0]
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print("width: ", w, "height: ", h)

plt.imshow(image)
```

```python
edges = (
    (0, 1),
    (1, 2),
    (2, 3),
    (3, 0),
    (4, 5),
    (5, 6),
    (6, 7),
    (7, 4),
    (0, 4),
    (1, 5),
    (2, 6),
    (3, 7)
)

def drawpoint(image, points, center):
    img_copy = image.copy()
    for i in points:
        cv2.circle(img_copy, tuple(i), 0, (255,0,0), 50)
    
    cv2.circle(img_copy, tuple(center), 0, (0,255,0), 50)
    plt.imshow(img_copy)

def drawline(image, points):
    img_copy = image.copy()
    for i in edges:
        x = points[i[0]]
        y = points[i[1]]
        cv2.line(img_copy, (x[0], x[1]) , (y[0], y[1]), (255,0,0), 10)
    plt.imshow(img_copy)
```

```python
drawpoint(image, data[7]["projection"][:8], (0,0))
```

```python
drawline(image, data[7]["projection"][0:8])
```
